```python
import pandas as pd
```

```python
df = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])
```

```python
df = pd.DataFrame(
[[4, 7, 10],
[5, 8, 11],
[6, 9, 12]],
index=[1, 2, 3],
columns=['a', 'b', 'c'])

display(df)
```

```python
df = pd.DataFrame(
{"variable" : [4 ,5, 6],
"value" : [210, 120,250],
"c" : [10, 11, 12]},
index = pd.MultiIndex.from_tuples(
[('d',1),('d',2),('e',2)],
names=['n','v']))

display(df)
```

```python
df = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [210, 120,250],
"c" : [10, 11, 12]},
index = [1, 2, 3])
display(df)
df = pd.melt(df)

display(df)
```

```python
df1 = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [210, 120,250],
"c" : [10, 11, 12]},
index = [1, 2, 3])

display(df1)
df2 = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [210, 120,250],
"c" : [10, 11, 12]},
index = [10, 11, 12])

display(df2)

df3=pd.concat([df1,df2])
display(df3)
df3=df3.reset_index()
display(df3)
```

```python
df1 = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [210, 120,250],
"c" : [10, 11, 12]},
index = [1, 2, 3])

display(df1)
df2 = pd.DataFrame(
{"d" : [4 ,5, 6],
"e" : [210, 120,250],
"f" : [10, 11, 12]},
index = [10, 11, 12])

display(df2)

df3=pd.concat([df1,df2])
display(df3)
```

```python
df1 = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [210, 120,250],
"c" : [10, 11, 12]},
index = [1, 2, 3])

display(df1)
df2 = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [210, 120,250],
"c" : [10, 11, 12]},
index = [10, 11, 12])

display(df2)

df3=pd.concat([df1,df2],axis=1)
display(df3)
```

```python
df1 = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [210, 120,250],
"c" : [10, 11, 12]},
index = [1, 2, 3])

df1=df1.drop(columns=['b'])
display(df1)
```
