```python
def dikdortgen(satir,sutun):
    i=0
    while(i<satir):
        t=0
        while(t<sutun):
            print("*",end="")
            t+=1
        print()
        i+=1

def basamak_toplami(sayi):
    s=str(sayi)

    t=0
    for i in s:
        t+=int(i)

    return t

def eslestir(s1,s2):
    t=0
    for i in range(min(len(s1),len(s2))):
        if(s1[i]==s2[i]):
            t+=1
    return t

def sirali_mi(liste):
    if(len(liste)<2):
        return True
    for i in range(len(liste)-1):
        if(liste[i]>liste[i+1]):
            return False
    return True

```

```python

liste=[3,5,9]

print(sirali_mi(liste))
```

```python
def faktoriyel(t):
    f=1
    i=1
    while(i<=t):
        f*=i
        i+=1

    return f

faktoriyel(100)
```

```python
def R_faktoriyel(t):
    if(t==1):
        return 1
    return R_faktoriyel(t-1)*t

R_faktoriyel(100)
```

```python
def r_ussu(a,b):
    if(b==0):
        return 1
    if(b==1):
        return a

    return r_ussu(a,b-1)*a

r_ussu(5,3)
```

```python
def r_fib(sayi):
    if(sayi<0):
        return -r_fib(-sayi)
    if(sayi==1 or sayi==2):
        return 1
    return r_fib(sayi-1)+r_fib(sayi-2)

```
